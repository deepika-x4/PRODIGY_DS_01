```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(file_path, skiprows=4)
pop_2024 = df[['Country Name', '2024']].dropna()
top10 = pop_2024.sort_values(by='2024', ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['2024'] / 1_000_000, color='skyblue', edgecolor='black')
plt.title("Top 10 Most Populous Countries (2024)", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population (millions)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
